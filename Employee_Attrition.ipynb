#### Step 4 - Run a Random Forest Model to derive the Importance Features to determine the Target Variable 

```python
# Converting String columns into integers

from sklearn.preprocessing import LabelEncoder
for column in employee_data.columns:
    if employee_data[column].dtype==np.number:
        continue
    else:
        employee_data[column]=LabelEncoder().fit_transform(employee_data[column])
```

```python
employee_data.head()
```

```python
# Building model by creating test and training data sets

X=employee_data.drop('Yes',axis=1)
```

```python
y=np.array(employee_data['Yes'])
```

```python
from sklearn.model_selection import train_test_split

# Set seed.
np.random.seed(2)

# Split the training and test sets. 
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.3)
```

```python
# Building a random forest model using all default parameters for now.

from sklearn.ensemble import RandomForestClassifier

ex_forest=RandomForestClassifier(
    criterion='gini', n_estimators=100, random_state=1)
```

```python
# Fit the model to the training data.

ex_forest.fit(X_train,y_train)
```

```python
# Predict on test data. 
ex_y_predict_forest=ex_forest.predict(X_test)

# Look at the first few predictions.
print(ex_y_predict_forest[0:5,])
```

```python
# Create a confusion matrix and calculate accuracy score for the ex_forest model.

from sklearn import metrics

ex_conf_matrix_test=metrics.confusion_matrix(y_test,ex_y_predict_forest)
```

```python
ex_test_accuracy_forest=metrics.accuracy_score(y_test,ex_y_predict_forest)
```

```python
print(ex_conf_matrix_test)
```

```python
print("Test Accuracy:",ex_test_accuracy_forest)
```
